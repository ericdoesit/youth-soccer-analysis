# Notebook 1: Data Loading & Validation
Load all five processed CSVs, inspect shapes and types, flag missing values or outliers.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

DATA = '../data/processed/'

programs = pd.read_csv(DATA + 'programs.csv')
costs    = pd.read_csv(DATA + 'costs.csv')
schols   = pd.read_csv(DATA + 'scholarship_outcomes.csv')
pros     = pd.read_csv(DATA + 'pro_outcomes.csv')
demo     = pd.read_csv(DATA + 'demographics.csv')

for name, df in [('programs', programs), ('costs', costs),
                 ('scholarships', schols), ('pro_outcomes', pros),
                 ('demographics', demo)]:
    print(f'\n=== {name} ===')
    print(f'  Shape: {df.shape}')
    print(f'  Columns: {list(df.columns)}')
    print(df.describe(include='all').T[['count','mean','min','max']].dropna(how='all'))
```

```python
# Flag any nulls
for name, df in [('programs', programs), ('costs', costs),
                 ('scholarships', schols), ('pro_outcomes', pros),
                 ('demographics', demo)]:
    nulls = df.isnull().sum()
    if nulls.any():
        print(f'\nNulls in {name}:')
        print(nulls[nulls > 0])
```

```python
# Confirm cost totals are internally consistent
costs['computed_total_mid'] = (
    costs['annual_club_fee_mid'] + costs['equipment_est'] +
    costs['tournament_fees_local'] + costs['tournament_fees_regional'] +
    costs['tournament_fees_national'] + costs['travel_est_local'] +
    costs['travel_est_regional'] + costs['travel_est_national'] +
    costs['coaching_extra'] + costs['camp_fees'] + costs['college_id_camps']
)
costs['total_check_diff'] = costs['computed_total_mid'] - costs['total_mid']
print('Cost total mid discrepancies (should be 0):')
print(costs[['cost_id', 'total_mid', 'computed_total_mid', 'total_check_diff']])
```
